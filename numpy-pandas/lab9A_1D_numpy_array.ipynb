## Task 3: use 1D numpy array to represent a point in 2D (2 point)

````python
p1=np.array([0.1, 0.2])
p2=np.array([10, 20])
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'o')
````

Compute the distance between two points
```
point1 is (x1, y1)
point2 is (x2, y2)
distance= square root of (x1-x2)**2 + (y1-y2)**2
use np.sqrt(a) to get the square root of a
```

````python
# your code (one line) to compute the distance between the two points using the above equation
# use p1[0], p[1], p2[0], and p2[1] in your code
distance = np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
distance
````

````python
# your code (one line) to compute the distance between the two points using the above equation
# do NOT use p1[n], p2[n] (n=0, 1) in your code
# use sqrt and sum
distance = np.sqrt(sum((a - b)**2 for a, b in zip(p1, p2)))
distance
````
