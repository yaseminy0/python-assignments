### Task 1 (10 points)

Implement a function according to the specification:
```python
def draw(t, x1, x2, color, fontsize, show_grid):
    # parameters:
    #   t is a 1D array (e.g. timepoints t=np.linspace(a, b, 100))
    #   x1 is a 1D array (e.g. x1 is sin(t))
    #   x2 is a 1D array (e.g. x2 is cos(t))
    #   color is the plot color
    #   fontsize is the fontsize of x_label, y_label, and title
    #   show_grid is True or False
    #
    # specification:
    #   compute the average of the two arrays: x = (x1+x2)/2
    #   plot the new array x
    #   set x_label to 't [time]'
    #   set y_label to 'x [signal amplitude]'
    #   set title to 'x(t) = (x1(t) + x2(t))/2'
    #   set color and fontsize
    #   show the grid in plot if show_grid is True
```

````python
#implement the function (5 points)
def draw(t, x1, x2, color, fontsize, show_grid):
    # compute the average of the two arrays
    x = (x1 + x2) / 2

    # plot the result
    plt.plot(t, x, color=color)

    plt.xlabel('t [time]', fontsize=fontsize)
    plt.ylabel('x [signal amplitude]', fontsize=fontsize)
    plt.title('x(t) = (x1(t) + x2(t))/2', fontsize=fontsize)

    if show_grid:
        plt.grid(True)

    plt.show()
````

````python
#test the function (5 points)
#your code:
t = np.linspace(0, 2*np.pi, 100)
x1 = np.sin(t)
x2 = np.cos(t)

draw(t, x1, x2, color='red', fontsize=12, show_grid=True)
````

### Task 2 (5 points)
Draw a triangle
```
This matrix (nested list) defines a triangle.
[[-0.5, 0],
 [ 0.5, 0],
 [ 0,   1]]
The above matrix has three rows
row-0 is point0, [x0, y0]
row-1 is point1, [x1, y1]
row-2 is point2, [x2, y2]
```

````python
#step 1 (1 point): create a numpy array to represent the triangle, name it Triangle
#do NOT define point0, point1, or point2
import numpy as np

Triangle = np.array([
    [-0.5, 0],
    [0.5, 0],
    [0, 1]
], dtype='float32')
````

````python
Triangle
````

````python
#step 2A (1 point) show the three points, using ax.plot
#complete this step using three lines of code (do not use ;), otherwise you get zero point for this step
fig, ax = plt.subplots()
ax.plot(Triangle[0, 0], Triangle[0, 1], 'ro')
ax.plot(Triangle[1, 0], Triangle[1, 1], 'ro')
ax.plot(Triangle[2, 0], Triangle[2, 1], 'ro')
plt.show()
````

````python
#step 2B (1 point): show the three points using ax.scatter
#complete this step within three lines of code (do not use ;), otherwise you get zero point for this step
fig, ax = plt.subplots()
ax.scatter(Triangle[0, 0], Triangle[0, 1], color='red')
ax.scatter(Triangle[1, 0], Triangle[1, 1], color='red')
ax.scatter(Triangle[2, 0], Triangle[2, 1], color='red')
plt.show()
````

````python
#step 3 (2 points): connect the three points by red lines  [ax.plot]
#complete this step within three lines of code (do not use ;), otherwise you get zero points for this step
fig, ax = plt.subplots()
ax.plot([Triangle[0,0], Triangle[1,0]], [Triangle[0,1], Triangle[1,1]], 'r',)
ax.plot([Triangle[1,0], Triangle[2,0]], [Triangle[1,1], Triangle[2,1]], 'r')
ax.plot([Triangle[2,0], Triangle[0,0]], [Triangle[2,1], Triangle[0,1]], 'r')
plt.show()
````

## Task 3 (5 points)
show histogram and bar chart of the customer data

````python
import pandas as pd
dataset=pd.read_csv('Customers.csv') # read a table from the file
dataset.head() # show the first five rows of the table dataset
````

````python
income=dataset['income'].values
income[0:4]
````

````python
spend=dataset['spend'].values
spend[0:4]
````

````python
#(1 point)
#show the histogram of spend and the histogram of income in two regions of the same figure
#blue color for the histogram of spend
#red color for the histogram of income
#set title of each regoin
#set bins=10
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.hist(dataset['spend'], bins=10, color='blue')
ax1.set_title('Spend')

ax2.hist(dataset['income'], bins=10, color='red')
ax2.set_title('Income')

plt.tight_layout()
plt.show()
````

````python
#(0.5 point)
#calculate the average income 
avg_income = dataset['income'].mean()
print(f"average_income is {avg_income}")
````

````python
#(0.5 point)
#calculate the average spend
avg_spend = dataset['spend'].mean()
print(f"average_spend is {avg_spend}")
````

````python
#(3 points)
#plot a bar chart
#read https://matplotlib.org/stable/gallery/lines_bars_and_markers/barh.html
import matplotlib.pyplot as plt

categories = ['average income', 'average spend']
values = [avg_income, avg_spend]

plt.barh(categories, values)
plt.xlabel('Value')
plt.title('Average Income vs Average Spend')
plt.show()
````
